```python
# DBSCAN-Based Hotspot Detection
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import geopandas as gpd
from shapely.geometry import Point

# ===========================
# 📌 1️⃣ LOAD DATASET
# ===========================
df = pd.read_csv("cleaned_fused_methane_climate_industry_data.csv")

# Drop rows with missing latitude/longitude
df = df.dropna(subset=["Latitude", "Longitude"])

# Convert latitude and longitude into NumPy array
coords = df[['Latitude', 'Longitude']].to_numpy()

# ===========================
# 📌 2️⃣ APPLY DBSCAN (ADJUSTED EPS)
# ===========================

# Set DBSCAN parameters for 5-6 clusters
db = DBSCAN(eps=0.05, min_samples=5, metric='haversine').fit(np.radians(coords))

# Add cluster labels to DataFrame
df['cluster'] = db.labels_

# ===========================
# 📌 3️⃣ PLOT UPDATED HOTSPOTS
# ===========================

plt.figure(figsize=(12, 7))

# Get unique clusters
unique_clusters = set(df["cluster"])

# Assign colors
colors = plt.cm.get_cmap("tab10", len(unique_clusters))

for cluster in unique_clusters:
    cluster_data = df[df["cluster"] == cluster]
    
    if cluster == -1:
        # Noise points (outliers) in black
        plt.scatter(cluster_data["Longitude"], cluster_data["Latitude"], c="black", label="Noise", alpha=0.6, edgecolors="k")
    else:
        # Cluster points in different colors
        plt.scatter(cluster_data["Longitude"], cluster_data["Latitude"], c=[colors(cluster)], label=f"Cluster {cluster}", alpha=0.7, edgecolors="k")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Updated DBSCAN Hotspot Detection (5-6 Clusters)")
plt.legend()
plt.grid(True)

# Save the plot
plt.savefig("dbscan_hotspots_optimized.png", dpi=300, bbox_inches="tight")

# Show the plot
plt.show()

# ===========================
# 📌 4️⃣ PRINT CLUSTER STATISTICS
# ===========================

# Count number of points in each cluster
cluster_counts = df['cluster'].value_counts()

print("\n📊  DBSCAN Clustering Results:\n")
print(cluster_counts)

# Save results
df.to_csv("dbscan_clustered_results_optimized.csv", index=False)
```

# Methane Emission Distribution

```python
import folium
from folium.plugins import TimestampedGeoJson
import pandas as pd

# Convert time column
df["Time"] = pd.to_datetime(df["Time"])
df["YearMonth"] = df["Time"].dt.to_period("M")  # Extract Year-Month

# Get unique time periods
time_periods = sorted(df["YearMonth"].unique())

# Create a base map
map_time = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=5)

# Prepare features for TimestampedGeoJson
features = []
for period in time_periods:
    sub_df = df[df["YearMonth"] == period]
    
    for _, row in sub_df.iterrows():
        feature = {
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": [row["Longitude"], row["Latitude"]],
            },
            "properties": {
                "time": row["Time"].isoformat(),  # ISO format timestamp
                "style": {"color": "red" if row["Methane_Concentration"] > df["Methane_Concentration"].quantile(0.75) else "blue"},
                "icon": "circle",
                "popup": f"<b>Methane:</b> {row['Methane_Concentration']:.2f} ppb<br><b>Date:</b> {row['Time']}"
            },
        }
        features.append(feature)

# Add timestamped data to the map
TimestampedGeoJson(
    {"type": "FeatureCollection", "features": features},
    period="P1M",  # Monthly intervals
    add_last_point=True,
    auto_play=True,
    loop=True,
    max_speed=1,
).add_to(map_time)

# Save map as HTML
html_filename = "methane_trend_over_time_2020_2024.html"
map_time.save(html_filename)

# Print confirmation message
print(f"✅ Interactive map saved as '{html_filename}'. Open it manually in a browser.")
```
